```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("Student_Performance.csv")

# 1. Convert score columns into NumPy arrays
# The column names were incorrect, I am using the correct ones found in df.columns: 'math_score', 'science_score', 'english_score'
scores = df[['math_score', 'science_score', 'english_score']].to_numpy()

# Separate arrays for clarity
math_scores = scores[:, 0]
science_scores = scores[:, 1]
english_scores = scores[:, 2]

# 2. Compute mean, median, standard deviation
print("Math Score Statistics:")
print("Mean:", np.mean(math_scores))
print("Median:", np.median(math_scores))
print("Standard Deviation:", np.std(math_scores), "\n")

print("Science Score Statistics:")
print("Mean:", np.mean(science_scores))
print("Median:", np.median(science_scores))
print("Standard Deviation:", np.std(science_scores), "\n")

print("English Score Statistics:")
print("Mean:", np.mean(english_scores))
print("Median:", np.median(english_scores))
print("Standard Deviation:", np.std(english_scores))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
df = pd.read_csv("Student_Performance.csv")

# Extract score columns
math_scores = df['math_score']
science_scores = df['science_score']
english_scores = df['english_score']

# 1. Histogram of Math Scores
plt.figure()
plt.hist(math_scores, bins=10)
plt.xlabel("Math Scores")
plt.ylabel("Number of Students")
plt.title("Histogram of Math Scores")
plt.show()

# 2. Bar Chart of Average Scores
avg_scores = [
    np.mean(math_scores),
    np.mean(science_scores),
    np.mean(english_scores)
]

subjects = ['Math', 'Science', 'English']

plt.figure()
plt.bar(subjects, avg_scores)
plt.xlabel("Subjects")
plt.ylabel("Average Score")
plt.title("Average Scores of Students")
plt.show()

# 3. Scatter Plot between Math and English Scores
plt.figure()
plt.scatter(math_scores, english_scores)
plt.xlabel("Math Scores")
plt.ylabel("English Scores")
plt.title("Math vs English Scores")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Student_Performance.csv")

# 1. Group scores by Gender
gender_group = df.groupby('gender')[['math_score', 'science_score', 'english_score']].mean()
print("Average Scores by Gender:")
print(gender_group, "\n")

# 2. Group scores by Parental Education
parent_edu_group = df.groupby('parent_education')[['math_score', 'science_score', 'english_score']].mean()
print("Average Scores by Parental Education:")
print(parent_edu_group, "\n")

# 3. Visualization: Gender-wise average scores
gender_group.plot(kind='bar')
plt.title("Average Scores by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Score")
plt.show()

# 4. Visualization: Parental Education-wise average scores
parent_edu_group.plot(kind='bar')
plt.title("Average Scores by Parental Education")
plt.xlabel("Parental Education Level")
plt.ylabel("Average Score")
plt.show()
```
